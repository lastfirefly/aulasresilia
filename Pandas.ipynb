##Criando um novo dataframe a partir de filtragem


Para criar um novo dataframe com as informações que você gostaria de localizar, pode-se ser utilizado o método de desigualdade, onde é verificado através de valores booleanos, se a condição está sendo respeitada, e depois disso, atribuindo a um novo dataframe em que todas as outras informações atribuídas às colunas filtradas, irão aparecer, por exemplo:

```python
df1 = df[df["a"]>=20]

df1
```

Um exemplo melhor ilustrado:

```python
tabela = pd.DataFrame({'rosa':[11,21,7],'roxo':[21,22,23], 'verde': [35, 20, 5], 'azul': [31, 21, 7]})

tabela['verde']>=20
```

```python
tabelaVerdes = tabela[tabela['verde']>=20]

tabelaVerdes
```
